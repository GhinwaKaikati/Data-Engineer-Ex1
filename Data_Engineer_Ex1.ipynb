```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/Colab Notebooks/UN ESCWA - Data Engineer/sample_data.csv'

dataset = pd.read_csv(file_path)
```

```python
dataset.fillna(value = 0, inplace = True)
```

```python
dataset['GeoAreaName'] = dataset['GeoAreaName'].str.rstrip()
```

```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/Colab Notebooks/UN ESCWA - Data Engineer/country_info.xlsx'

country_info = pd.read_excel(file_path)
```

```python
print("Dataset columns:", dataset.columns.tolist())
print("Country info columns:", country_info.columns.tolist())

country_info = country_info.rename(columns={'ISO3': 'ISO', 'Arabic Name': 'GeoAreaName_ar'})

dataset = dataset.merge(
    country_info[['Country', 'ISO', 'GeoAreaName_ar']],
    left_on='GeoAreaName',
    right_on='Country',
    how='left'
)

if 'Country' in dataset.columns and 'Country' != 'GeoAreaName':
    dataset = dataset.drop('Country', axis=1)


dataset.to_csv('updated_dataset.csv', index=False)
```

```python
dataset
```

```python
unmatched_rows = dataset[dataset['ISO'].isna()].copy()
print(f"Number of rows without matching ISO codes: {len(unmatched_rows)}")
print(f"Original row indices of unmatched rows: {unmatched_rows.index.tolist()}")
```

```python
empty_columns = [col for col in dataset.columns if dataset[col].isna().all()]
print(f"Columns with no data: {empty_columns}")
```

```python
dataset_cleaned = dataset.dropna(axis=1, how='all')

print(f"Original DataFrame had {len(dataset.columns)} columns")
print(f"Cleaned DataFrame has {len(dataset_cleaned.columns)} columns")
print(f"Dropped {len(dataset.columns) - len(dataset_cleaned.columns)} empty columns")

dataset_cleaned.to_json('output_data.json', orient='records')
```
